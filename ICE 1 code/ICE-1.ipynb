```python
import urllib.request

from bs4 import BeautifulSoup

import nltk

from nltk.corpus import stopwords

import matplotlib.pyplot as plt

response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')

html = response.read() 

soup = BeautifulSoup(html,"html5lib")

text = soup.get_text(strip=True)

tokens = [t for t in text.split()]  # splitting the words into the list


freq1 = nltk.FreqDist(tokens)

mylst1 = []

for key,val in freq1.items():
    
    if(val < 5 or (key.isdigit())):  # Removing the keys with values less than 5 and Removing the keys which are numerical
        mylst1.append(key)
        
for i in mylst1:
    if(i in freq1):
        del freq1[i]

plt.title("Frequency Distribution with stopwords")
freq1.plot(10, cumulative = False)  #plotting frequency distribution with stopwords

clean_tokens = tokens[:]

sr = stopwords.words('english')

for token in tokens:
    if token in sr:
        clean_tokens.remove(token)  #removing stopwords
        
freq2 = nltk.FreqDist(clean_tokens)

mylst2 = []

for key,val in freq2.items():
    
    if(val < 5 or (key.isdigit())):  # Removing the keys with values less than 5 and Removing the keys which are numerical
        mylst2.append(key)
        
for i in mylst2:
    if(i in freq2):
        del freq2[i]

plt.title("Frequency Distribution without stopwords")        
freq2.plot(10, cumulative = False)   #plotting frequency distribution with stopwords
```
